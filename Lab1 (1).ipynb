```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotnine as pln
dataset1=pd.read_csv('/content/q1data (3).csv',
                     delimiter=',')
dataset1
```

```python
dataset1.info()
```

```python
from plotnine.geoms.annotate import annotate
from plotnine import ggplot, geom_point, aes, geom_boxplot, labs, scale_size, theme_linedraw, theme, scale_color_manual
(ggplot(dataset1)
      + aes(x = "income", y = "life_exp",color="four_regions")
   +annotate("text", x = 60000, y = 50, label = "2010", size = 100, colour = "grey", alpha = 0.5)
+ annotate("text", x = 85000, y = 20, label = "GDP per capital (Price and inflation adjusted, in PP $2017)", size = 5, colour = "black")
      + geom_point(aes(size="population"))
      + scale_size(range=(1,20))
      + labs(title = "Income and population by country",
             caption = "Data for 2010",
             x = "Incom",
             y = "Life expectancy")

      + theme_linedraw()
      + theme(figure_size=(12, 6),)
      + scale_color_manual(values = ("blue", "green","pink","orange"),name = "Continents")
)
```

from my point of view, this geometry (geom_point) is the most successful for visualizing four-dimensional data. Other geometries are convenient to use for lower-dimensional data. We use this geometry to visualize 5-dimensional data (just for sample). To do this, we use the aesthetic - shape. The above is true for tasks 2 and 3.

```python
from plotnine import ggplot, geom_point, aes, geom_boxplot, labs, scale_size, theme_linedraw, scale_color_manual
(ggplot(dataset1)
      + aes(x = "income", y = "life_exp",color="four_regions",shape="eight_regions")
      + geom_point(aes(size="population"))
      + scale_size(range=(1,20))
      + labs(title = "Income and population by country",
             caption = "Data for 2010",
             x = "Incom",
             y = "Life expectancy")
      + theme_linedraw()
      + scale_color_manual(values = ("blue", "green","pink","orange"),name = "Continents")
)
```

```python
dataset2=pd.read_csv('/content/q2data (2).csv',
                     delimiter=',')
dataset2
```

```python
dataset2.info()
```

We are going to use dataset2.['energy'] as a scale (size) of points for a scatterplot. But it is represented in object format. It will not allow us to use the data for the scaling of a point. So, we need to perform a transformation of this data in number format.

```python
dataset2['energy'] = pd.to_numeric(dataset2['energy'],errors='coerce')
dataset2.info()
```

Now we can see the float64 format of 'energy'

Visualization of dataset2

```python
(ggplot(dataset2)
      + aes(x = "exports",y = "imports",color="four_regions")
      + annotate("text", x = 100, y = 80, label = "1997", size = 90, colour = "grey", alpha = 0.5)

      + geom_point(aes(size="energy"),alpha=1/4)

      + scale_size(range=(1,10))
      + labs(title = "Import/export and energy",
             caption = "Data for 1997",
             x = "Export % of GDP",
             y = "Import % of GDP",
            fill = "four_regions",
        )

      + theme_linedraw()
      + scale_color_manual(values = ("blue", "green","pink","orange"),name = "Continents")
)
```
